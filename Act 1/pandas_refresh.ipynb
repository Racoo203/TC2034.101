# Eliminar filas o columnas

La funcion `df.drop( fila/columna, axis=0/1, inplace=True/False )` sirve para eliminar filas o columnas. El primer parametro es la `fila` o `columna` que queremos eliminar. El segundo es el `axis`, puede ser `0` (para filas) o `1` (para columnas). El tercer parametro inplace puede ser `True` (si la modificion se hara directo sobre `df`) o `False` (si no). En caso de que no, `drop` regresara un nuevo dataframe modificado.


```python
# se elimina la fila que tenga la posicion 1, directo sobre df
#
df.drop(df.index [ [1] ], axis=0, inplace=True)
df.head()

#   se elimina la columna 3, pero el resultado se relejara en df2 y no en df
# df2 = df.drop(3, axis=1, inplace=False)
df2 = df.drop(df.columns[3], axis=1, inplace=False)

df2.head()
df.head()
```

# Manejo de NaN

Algunas columnas tienen valores faltantes (NaN). Por ejemplo, la columna **Critic_Score** tiene celdas con `NaN`. Se puede averiguar cuales datos de **Critic_Score** usando la funcion `isna`. El resultado es una serie de booleanos, donde `True` indica cuales son `NaN`.  La funcion `isnull` es similar a `isna`.

```python
df['Critic_Score'].isna()
```

Es usual elminar los NaN.  La funcion `drop` puede hacerlo si recibe los indices que cumplen la condicion de ser NaN.

```python
df.drop( df[ df['Critic_Score'].isna() ].index  )
```

La funcion `fillna` sustituye los `NaN` por otra cosa que se indique. Esta es su sintaxis: `df['columna'] = df['columna'].fillna('valor sustituto')`   

La funcion mean sirve para obtener el promedio. Se puede usar sobre una columna `df['NA_Sales'].mean()`. Por default, `mean` omitira los NaN que existan.

```python
df['NA_Sales'].mean()
```
